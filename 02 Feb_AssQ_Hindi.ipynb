```python
#answer 1
The for loop is used when we already know the number of iterations, which means when we know how many times a statement has to be
executed. That is why we have to specify the ending point in the for loop initialization.

When we need to end the loop on a condition other than the number of times, we use a while loop. In this case, it is not necessary
to know the condition beforehand. That is why we can give a boolean expression in the initialization of the loop.
# while loop EXP:
number=int(input("Enter any number to devide by 3:")),
num=number
j=0
while number>10:
    number=number/3
    j=j+1
print("Number:"+str(num)+" is :"+str(j)+" times divided by 3 before 11") 

#for loop example:
for i in range(10):
    if i % 2 == 0:
        print(i)
```

```python
#answer 3
mon_unit=int(input("Please enter your monthly unit consumed of electricity: "))
amountpay=0

if mon_unit<=100:
    amountpay=mon_unit*4.5
elif mon_unit<=200:
    amountpay=(100*4.5)+((mon_unit-100)*6)
elif mon_unit<=300:
    amountpay=(100*4.5)+(100*6)+((mon_unit-200)*10)
else:
    amountpay=(100*4.5)+(100*6)+(100*10)+((mon_unit-300)*20)

print("Total electricity bill is: ",amountpay)
```
