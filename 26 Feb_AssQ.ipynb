```python
#Q1. Is there any difference in the data type of variables list_ and array_list? If there is then write a code to print the data types of both the variables.
"""
1.	The list can store the value of different types.	Array-It can only consist of value of same type.
2.	The list cannot handle the direct arithmetic operations. Array-	It can directly handle arithmetic operations.
3.	We need to import the array before work with the array.	The lists are the build-in data structure so we don't need to import it.
"""
import numpy as np
list_ = [ '1' , '2' , '3' , '4' , '5' ]
array_list = np.array(object = list_)
print(type(list_))
print(type(array_list))
```

```python
#Q2. Write a code to print the data type of each and every element of both the variables list_ and arra_list.
for i in list_:
    print(i," ",type(i))
for i in array_list:
    print(i," ",type(i))
```

```python
#3 Will there be any difference in the data type of the elements present in both the variables, list_ and arra_list? If so then print the data types of each and every element present in both the variables, list_
#and arra_list.
array_list = np.array(object = list_, dtype = int)
for i in list_:
    print(i," ",type(i))
for i in array_list:
    print(i," ",type(i))
```
