```python
#Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
"""
In MongoDB, sorting is done by the sort() method. The sort() method consists of two basic building blocks.
These building blocks are fields to be sorted and the sort order.
The sorting order in MongoDB is defined by either a one (1) or a minus (-1). Here the positive one represents the 
ascending order, while the negative one represents the descending order.
 """  
data2=[{"rollno":1,"name":"garvi"},
       {"rollno":2,"name":"anya"},
       {"rollno":3,"name":"vian"},
       {"rollno":4,"name":"tanvi"}]
student=db["my_student"]
student.insert_many(data2)

for i in student.find().sort('rollno', pymongo.ASCENDING):
    print(i)
```

```python
#Q7. Explain why delete_one(), delete_many(), and drop() is used.
student.delete_one({"rollno":1})
for i in student.find().sort('rollno', pymongo.ASCENDING):
    print(i)
```
