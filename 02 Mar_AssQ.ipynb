```python
"""
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
Note: Also add title, xlabel, and ylabel to the plot.
"""
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random number")
```

```python
"""
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
subplot divides the current figure into rectangular panes that are numbered row-wise. Each pane contains an axes. Subsequent plots are output to the current pane.
subplot(m,n,p) creates an axes in the p-th pane of a figure divided into an m-by-n matrix of rectangular panes.
"""
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
a = np.array([0, 100, 200, 300, 400, 500])
b = np.array([50, 20, 40, 20, 60, 70])
c = np.array([10, 20, 30, 40, 50, 60])
d = np.array([200, 350, 250, 550, 450, 150])

plt.plot(x, a, label = "line 1")
plt.plot(x, b, label = "line 2")
plt.plot(x, c, label = "line 3")
plt.plot(x, d, label = "line 4")
plt.legend()
plt.show()
```
