```python
#Q3. Which Python statements are used to catch and handle exceptions? Explain with an example
"""
The try and except block in Python is used to catch and handle exceptions. 
Python executes code following the try statement as a “normal” part of the program. The code that follows the except statement is the program's 
response to any exceptions in the preceding try clause.
"""
try:
    a=10/0
except ZeroDivisionError as e:
    print(e)
```

```python
# Q6. Create a custom exception class. Use this class to handle an exception.

class validateage(Exception):
    def __init__(self,msg):
        self.msg=msg
        
def validaeage(age):
    if age<0:
        raise validateage("entered age is negative")
    elif age>200:
        raise validateage("entered age is very high")
    else:
        print("age is valid")
        
try:
    age=int(input("enter your age:"))
    validaeage(age)
except validateage as e:
    print(e)
```
