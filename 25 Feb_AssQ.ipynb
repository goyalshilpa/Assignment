```python
#Q1. Write a code to print the data present in the second row of the dataframe, df.
import pandas as pd
course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2,3,6,4]
df = pd.DataFrame(data = {'course_name' : course_name, 'duration' : duration})
df.loc[[1]]
```

```python
#Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
"""
The main distinction between the two methods is:
loc gets rows (and/or columns) with particular labels.
iloc gets rows (and/or columns) at integer locations.
"""
df.loc[0:2,['PassengerId','Survived','Pclass']]
df.iloc[0:2,[0,1,2]]
```

```python
#Q6. What do you understand about the windows function in pandas and list the types of windows functions?
"""
Window functions allow us to perform an operation with a given row’s data and data from another row that is a specified number of rows away 
— this “number of rows away value” is called the window.
"""
df1['column_3'].expanding(2).sum()
df1['column_3'].rolling(3).sum()
```

```python
#Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type. 
#The program should prompt the user to enter the file path, column name, and category order, and then display the sorted data.
import pandas as pd
location=input("enter file location:")
column=input("enter column to be categorized:")
order=input("enter order for categorized:")
data=pd.read_csv(location)
if order=='True':
    cat=pd.Categorical(data[column],ordered=True)
else:
    cat=pd.Categorical(data[column],ordered=False)
cat
```
